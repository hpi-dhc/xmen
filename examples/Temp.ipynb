```python
from xmen import evaluation
from xmen.data import make_document, Entity, Concept
```

```python
gt = [
    make_document(
        [
            Entity([[11, 17]], "entity", concepts=[
                Concept("c1", db_name="UMLS"),
            ]),
        ]
    )
]
pred = [
    make_document(
        [
            Entity([[11, 17]], "entity", concepts=[
                Concept("c1", db_name="UMLS")
            ]),
            Entity([[11, 15]], "entity", concepts=[
                Concept("c1", db_name="UMLS")
            ]),
            Entity([[16, 17]], "entity", concepts=[
                Concept("c2", db_name="UMLS")
            ]),
        ]
    )
]
```
